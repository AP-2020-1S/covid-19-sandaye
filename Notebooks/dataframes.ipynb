```python
import pandas as pd
import numpy as np
import datetime
import requests
import os.path
import os
```

```python
dfcovid_ciudades = pd.read_csv(os.path.join('../Output', 'data_final.csv'))
dfcovid_ciudades.head()
```

```python
df1 = pd.to_datetime(dfcovid_ciudades['FECHA REPORTE WEB'])
dfcovid_confirmados = df1.groupby(df1.dt.floor('d')).size().reset_index(name='COUNT')
print (dfcovid_confirmados)
```

```python
df2 = pd.to_datetime(dfcovid_ciudades['FECHA RECUPERADO'])
dfcovid_recuperados = df2.groupby(df2.dt.floor('d')).size().reset_index(name='COUNT')
print (dfcovid_recuperados)
```

```python
df3 = pd.to_datetime(dfcovid_ciudades['FECHA DE MUERTE'])
dfcovid_muertes = df3.groupby(df3.dt.floor('d')).size().reset_index(name='COUNT')
print (dfcovid_muertes)
```
